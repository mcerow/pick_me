## Combine Results Functions

To get the data for combine performance, I built the following functions. This way, when I'm ready, I can pass in each function to append a list and later create a dictionary & dataframe. 

```python
def names(data):
    """
    This function gets the name of a player from html soup and adds each to a list.
    
    Parameters:
        data: html data from a given requested page.
        
    Returns:
        List of player names for a given webpage.
        
    """
    all_names = []
    count = 0
    for i in data:
        all_names.append(data[count].get_text())
        count += 1
    return all_names
```

```python
def get_position(data):
    """
    This function gets the position a player plays from html soup and adds each to a list.
    
    Parameters:
        data: html data from a given requested page.
        
    Returns:
        List of position for a given webpage.
        
    """
    all_positions = []
    count = 0
    for i in data:
        if count in range(len(data)):
            all_positions.append(data[count].get_text())
            count += 12
    
    return all_positions
```

```python
def get_school(data):
    """
    This function gets the name of the school a player attends from html soup and adds each to a list.
    
    Parameters:
        data: html data from a given requested page.
        
    Returns:
        List of schools for a given webpage.
        
    """
    all_schools = []
    count = 1
    for i in data:
        if count in range(len(data)):
            all_schools.append(data[count].get_text())
            count += 12
    
    return all_schools
```

```python
# this column will be used to then search for the college stats

def get_stats_links(data):
    """
    This function gets the link for college stats for a player from html soup and adds each to a list. The link is 
    later used 
    
    Parameters:
        data: html data from a given requested page.
        
    Returns:
        List of links associated with the college stats for a player for a given webpage.
        
    """
    stats_links = []
    count = 0
    for i in data:
        if count in range(len(data)):
            link = data[count]
            link = str(link)
            link = link.split('<')
            link = link[2]
            link = link.lstrip('a href =')
            link = link.rstrip('> College Stats')
            link = link.replace('"', '')
            stats_links.append(link)
            count += 9
        else:
            continue
    return stats_links
```

```python
def get_height(data):
    """
    This function gets the height for a player from html soup and adds each to a list.
    
    Parameters:
        data: html data from a given requested page.
        
    Returns:
        List of player heights for a given webpage.
        
    """
    all_heights = []
    count = 3
    for i in data:
        if count in range(len(data)):
            all_heights.append(data[count].get_text())
            count += 12
    
    return all_heights
```

```python
def get_weight(data):
    """
    This function gets the weight for a player from html soup and adds each to a list.
    
    Parameters:
        data: html data from a given requested page.
        
    Returns:
        List of player weights for a given webpage.
        
    """
    all_weights = []
    count = 4
    for i in data:
        if count in range(len(data)):
            all_weights.append(data[count].get_text())
            count += 12
    
    return all_weights
```

```python
def get_40yd(data):
    """
    This function gets results for the 40yd dash for a player from html soup and adds each to a list.
    
    Parameters:
        data: html data from a given requested page.
        
    Returns:
        List of results for the 40yd dash for a given webpage.
        
    """
    all_40yds = []
    count = 5
    for i in data:
        if count in range(len(data)):
            all_40yds.append(data[count].get_text())
            count += 12
    
    return all_40yds
```

```python
def get_vertical(data):
    """
    This function gets results for the vertical jump for a player from html soup and adds each to a list.
    
    Parameters:
        data: html data from a given requested page.
        
    Returns:
        List of results for the vertical jump for a given webpage.
        
    """
    all_verticals = []
    count = 6
    for i in data:
        if count in range(len(data)):
            all_verticals.append(data[count].get_text())
            count += 12
    
    return all_verticals
```

```python
def get_bench(data):
    """
    This function gets results for the bench press for a player from html soup and adds each to a list.
    
    Parameters:
        data: html data from a given requested page.
        
    Returns:
        List of results for the bench press for a given webpage.
    
    """
    all_bench = []
    count = 7
    for i in data:
        if count in range(len(data)):
            all_bench.append(data[count].get_text())
            count += 12
    
    return all_bench
```

```python
def get_broadjump(data):
    """
    This function gets results for the broad jump for a player from html soup and adds each to a list.
    
    Parameters:
        data: html data from a given requested page.
        
    Returns:
        List of results for the broad jump for a given webpage.
    
    """
    all_jumps = []
    count = 8
    for i in data:
        if count in range(len(data)):
            all_jumps.append(data[count].get_text())
            count += 12
    
    return all_jumps
```

```python
def get_3cone(data):
    """
    This function gets results for the 3 cone drill for a player from html soup and adds each to a list.
    
    Parameters:
        data: html data from a given requested page.
        
    Returns:
        List of results for the 3 cone drill for a given webpage.
    
    """
    all_cones = []
    count = 9
    for i in data:
        if count in range(len(data)):
            all_cones.append(data[count].get_text())
            count += 12
    
    return all_cones
```

```python
def get_shuttle(data):
    """
    This function gets results for the shuttle run for a player from html soup and adds each to a list.
    
    Parameters:
        data: html data from a given requested page.
        
    Returns:
        List of results for the shuttle run for a given webpage.
    
    """
    all_shuttles = []
    count = 10
    for i in data:
        if count in range(len(data)):
            all_shuttles.append(data[count].get_text())
            count += 12
    
    return all_shuttles
```

```python
def get_draft(data):
    """
    This function gets the draft information for a player from html soup and adds each to a list.
    
    Parameters:
        data: html data from a given requested page.
        
    Returns:
        List of draft results for a given webpage.
    
    """
    all_draft = []
    count = 11
    for i in data:
        if count in range(len(data)):
            all_draft.append(data[count].get_text())
            count += 12
    
    return all_draft
```

```python
# scraping pro-football-reference for data

def get_combine_history(links):
    """
    This function gets all the results for the combine and append each to the appropriate list. Each list is then 
    added to a dictionary to later be used to create a dataframe.
    
    Parameters:
        links: list of links to request to then parse through.
        
    Returns:
        Dictionary of all results for a list of links around combine data.
    
    """
    
    SearchLinks = []
    PlayerNames = []
    Positions = []
    Schools = []
    StatsLinks = []
    Heights = []
    Weights = []
    _40yds = []
    Verticals = []
    Benched = []
    BroadJumps = []
    _3cones = []
    Shuttles = []
    Drafts = []
    
    for link in tqdm(links):
        url = link
        html_page = requests.get(url) 
        soup = BeautifulSoup(html_page.content, 'html.parser')

        player_container = soup.find('tbody')
        player_name = player_container.findAll('th', scope='row')
        player_deets = player_container.findAll('td')
        stats_finder = player_container.findAll('td', {'class':'right'})
           
        for name in names(player_name):
            PlayerNames.append(name)
            SearchLinks.append(str(link))

        for position in get_position(player_deets):
            Positions.append(position)
     
        for school in get_school(player_deets):
            Schools.append(school)
        
        for statlink in get_stats_links(stats_finder):
            StatsLinks.append(statlink)

        for height in get_height(player_deets):
            Heights.append(height)

        for weight in get_weight(player_deets):
            Weights.append(weight)       

        for _40yd in get_40yd(player_deets):
            _40yds.append(_40yd)
            
        for vertical in get_vertical(player_deets):
            Verticals.append(vertical)

        for bench in get_bench(player_deets):
            Benched.append(bench)

        for jump in get_broadjump(player_deets):
            BroadJumps.append(jump)

        for cone in get_3cone(player_deets):
            _3cones.append(cone)      

        for shuttle in get_shuttle(player_deets):
            Shuttles.append(shuttle)

        for draft in get_draft(player_deets):
            Drafts.append(draft)

        time.sleep(.5)

    combine_dictionary = {'Link': SearchLinks, 'PlayerName': PlayerNames, 'Position': Positions, 'School': Schools, 
                          'CollegeStats': StatsLinks, 'Height': Heights, 'Weight': Weights, '40yd': _40yds, 
                          'Vertical': Verticals, 'Bench': Benched, 'BroadJump': BroadJumps, '3Cone': _3cones, 
                          'Shuttle': Shuttles, 'Drafted': Drafts}
    return combine_dictionary
```

## College Stats Functions

Like gathering data for the combine results, I will use functions to get the data from the college stats sites. The first one creates containers for each of the tables on a given page. sports-reference gets a little funky. After the first table, each table is actually captured in a comment. The function below looks to see if first the table under the designated header is in a comment, or a child of the division. The comment is then passed through as an html itself and pass this new html instance through a new soup. If that's the case, it will just create a normal container.

A huge thanks to [Mark von Oven](https://stackoverflow.com/questions/53543064/using-beautifulsoup-to-parse-a-big-comment) on StackOverflow for answering his own question and saving my all sorts of grief trying to access this data. I updated his code for my needs in this case.

```python
def defense_container(soup):
    """
    This function find the container where all defense information is stored.
    
    Parameters:
        soup: full html page that is searched through.
        
    Returns:
        Section of html that has the information for defense statistics.
    """
        
    # defense container
    try:
        all_defense = soup.find('div', id='all_defense')
        for item in all_defense.children:
            if isinstance(item, bs4.element.Comment):
                html = item
        defense = BeautifulSoup(html, features="html.parser")
    except:
        html = ''
        defense = soup.find('div', id='all_defense')
        
    return defense
```

```python
def passing_container(soup):
    """
    This function find the container where all passing information is stored.
    
    Parameters:
        soup: full html page that is searched through.
        
    Returns:
        Section of html that has the information for passing statistics.
    """
    # passing container
    try:
        all_passing = soup.find('div', id='all_passing')
        for item in all_passing.children:
            if isinstance(item, bs4.element.Comment):
                html = item
        passing = BeautifulSoup(html, features="html.parser")
    except:
        html = ''
        passing = soup.find('div', id='all_passing')
        
    return passing
```

```python
def rushing_container(soup):
    """
    This function find the container where all rushing information is stored.
    
    Parameters:
        soup: full html page that is searched through.
        
    Returns:
        Section of html that has the information for rushing statistics.
    """
    # rushing container
    try:
        all_rushing = soup.find('div', id='all_rushing')
        for item in all_rushing.children:
            if isinstance(item, bs4.element.Comment):
                html = item
        rushing = BeautifulSoup(html, features="html.parser")
    except:
        html = ''
        rushing = soup.find('div', id='all_rushing')
        
    return rushing
```

```python
def scoring_container(soup):
    """
    This function find the container where all scoring information is stored.
    
    Parameters:
        soup: full html page that is searched through.
        
    Returns:
        Section of html that has the information for scoring statistics.
    """   
    # scoring container
    try:
        all_scoring = soup.find('div', id='all_scoring')
        for item in all_scoring.children:
            if isinstance(item, bs4.element.Comment):
                html = item
        scoring = BeautifulSoup(html, features="html.parser")
    except:
        html = ''
        scoring = soup.find('div', id='all_scoring')  
        
    return scoring
```

```python
def kicking_container(soup):
    """
    This function find the container where all kicking information is stored.
    
    Parameters:
        soup: full html page that is searched through.
        
    Returns:
        Section of html that has the information for kicking statistics.
    """
    # punting container
    try:
        all_kicking = soup.find('div', id='all_kicking')
        for item in all_kicking.children:
            if isinstance(item, bs4.element.Comment):
                html = item
        kicking = BeautifulSoup(html, features="html.parser")
    except:
        html = ''
        kicking = soup.find('div', id='all_kicking')
        
    return kicking
```

```python
def punt_ret_container(soup):
    """
    This function find the container where all punt returning information is stored.
    
    Parameters:
        soup: full html page that is searched through.
        
    Returns:
        Section of html that has the information for punt returning statistics.
    """    
    # punt ret container
    try:
        all_punt_ret = soup.find('div', id='all_punt_ret')
        for item in all_punt_ret.children:
            if isinstance(item, bs4.element.Comment):
                html = item
        punt_ret = BeautifulSoup(html, features="html.parser")
    except:
        html = ''
        punt_ret = soup.find('div', id='all_punt_ret')
        
    return punt_ret
```

```python
# web scraper

def get_college_stats(link_list):
    """
    This function searches through a list of links and finds the necessary information around all college statistics
    interested in gathering for this project. It finds the stats around defense, passing, rushing, kicking, punt
    returning and scoring.
    
    Parameters:
        link_list: list of links for each individual player.
        
    Returns:
        Dataframe with all metrics appended.
    
    """
    
    for link in tqdm(link_list):
        url = link
        html_page = requests.get(url) 
        soup = BeautifulSoup(html_page.content, 'html.parser')
        
        # containers
        def_container = defense_container(soup)
        pass_container = passing_container(soup)
        rush_container = rushing_container(soup)
        scor_container= scoring_container(soup)
        kick_container = kicking_container(soup)
        punret_container = punt_ret_container(soup) 
        
        try:
            defense = def_container.find('tbody')
        except AttributeError:
            defense = defense_container(soup)
            
        try:
            passing = pass_container.find('tbody')
        except AttributeError:
            passing = passing_container(soup)
            
        try:
            rushing = rush_container.find('tbody')
        except AttributeError:
            rushing = rushing_container(soup)
            
        try:
            scoring = scor_container.find('tbody')
        except AttributeError:
            scoring = scoring_container(soup)
        
        try:
            kicking = kick_container.find('tbody')
        except AttributeError:
            kicking = kicking_container(soup)
        
        try:
            punt_ret = punret_container.find('tbody')
        except AttributeError:
            punt_ret = punt_ret_container(soup)
        
        # defense 
        # defense games
        try:
            def_games = defense.findAll('td', {'data-stat': 'g'})
            defense_games = def_games[-1].get_text()
                  
        except AttributeError:
            defense_games = 0
        
        # defense solo tackles
        try:
            solotackles = defense.findAll('td', {'data-stat': 'tackles_solo'})
            solotackles = solotackles[-1].get_text()

        except AttributeError:
            solotackles = 0
        
        # defense assisted tackles
        try:
            assistedtackles = defense.findAll('td', {'data-stat': 'tackles_assists'})
            assistedtackles = assistedtackles[-1].get_text()
        
        except AttributeError:
            assistedtackles = 0
            
        # defense tackles_total
        try:
            ttl_tackles = defense.findAll('td', {'data-stat': 'tackles_total'})
            ttl_tackles = ttl_tackles[-1].get_text()
        
        except AttributeError:
            ttl_tackles = 0
            
        # tackles for a loss
        try:
            tackle_loss = defense.findAll('td', {'data-stat': 'tackles_loss'})
            tackle_loss = tackle_loss[-1].get_text()
            
        except AttributeError:
            tackle_loss = 0
            
        # sacks
        try:
            sacks = defense.findAll('td', {'data-stat': 'sacks'})
            sacks = sacks[-1].get_text()
            
        except AttributeError:
            sacks = 0
            
        # defense interception
        try:
            defense_int = defense.findAll('td', {'data-stat': 'def_int'})
            defense_int = defense_int[-1].get_text()
            
        except AttributeError:
            defense_int = 0
            
        # defense interception yards
        try:
            int_yards = defense.findAll('td', {'data-stat': 'def_int_yds'})
            int_yards = int_yards[-1].get_text()
            
        except AttributeError:
            int_yards = 0
            
        # yards per interception
        try:
            yards_per_int = defense.findAll('td', {'data-stat': 'def_int_yds_per_int'})
            yards_per_int = yards_per_int[-1].get_text()
            
        except AttributeError:
            yards_per_int = 0
            
        # pick6
        try:
            pick6 = defense.findAll('td', {'data-stat': 'def_int_td'})
            pick6 = pick6[-1].get_text()
            
        except AttributeError:
            pick6 = 0
            
        # defended passes
        try:
            defended_pass = defense.findAll('td', {'data-stat': 'pass_defended'})
            defended_pass = defended_pass[-1].get_text()
            
        except AttributeError:
            defended_pass = 0
            
        # recovered fumbles
        try:
            recovered_fumbles = defense.findAll('td', {'data-stat': 'fumbles_rec'})
            recovered_fumbles = recovered_fumbles[-1].get_text()
            
        except AttributeError:
            recovered_fumbles = 0
            
        # fumble recovery yards
        try:
            fumble_recover_yrds = defense.findAll('td', {'data-stat': 'fumbles_rec_yds'})
            fumble_recover_yrds = fumble_recover_yrds[-1].get_text()
            
        except AttributeError:
            fumble_recover_yrds = 0

        # fumble recovery for a td
        try:
            recovered_fumbles_td = defense.findAll('td', {'data-stat': 'fumbles_rec_td'})
            recovered_fumbles_td = recovered_fumbles_td[-1].get_text()
            
        except AttributeError:
            recovered_fumbles_td = 0
            
        # forced fumbles
        try:
            forced_fumbles = defense.findAll('td', {'data-stat': 'fumbles_forced'})
            forced_fumbles = forced_fumbles[-1].get_text()
            
        except AttributeError:
            forced_fumbles = 0  
            
        # passing 
        # passing games
        try:
            passing_games = passing.findAll('td', {'data-stat': 'g'})
            passing_games = passing_games[-1].get_text()
            
        except AttributeError:
            passing_games = 0
           
        # completions
        try:
            completions = passing.findAll('td', {'data-stat': 'pass_cmp'})
            completions = completions[-1].get_text()
            
        except AttributeError:
            completions = 0    
            
        # attempts
        try:
            pass_attempts = passing.findAll('td', {'data-stat': 'pass_att'})
            pass_attempts = pass_attempts[-1].get_text()
            
        except AttributeError:
            pass_attempts = 0 
            
        # completion percentage
        try:
            complete_rate = passing.findAll('td', {'data-stat': 'pass_cmp_pct'})
            complete_rate = complete_rate[-1].get_text()
            
        except AttributeError:
            complete_rate = 0 

        # passing yards
        try:
            passing_yards = passing.findAll('td', {'data-stat': 'pass_yds'})
            passing_yards = passing_yards[-1].get_text()
            
        except AttributeError:
            passing_yards = 0 
            
        # yards per attempt
        try:
            yds_per_attempt = passing.findAll('td', {'data-stat': 'pass_yds_per_att'})
            yds_per_attempt = yds_per_attempt[-1].get_text()
            
        except AttributeError:
            yds_per_attempt = 0 
                            
        # adj yards per attempt
        try:
            adj_yds_per_attempt = passing.findAll('td', {'data-stat': 'adj_pass_yds_per_att'})
            adj_yds_per_attempt = adj_yds_per_attempt[-1].get_text()
            
        except AttributeError:
            adj_yds_per_attempt = 0 
        
        # pass tds
        try:
            pass_tds = passing.findAll('td', {'data-stat': 'pass_td'})
            pass_tds = pass_tds[-1].get_text()
            
        except AttributeError:
            pass_tds = 0 
                       
        # pass interceptions
        try:
            pass_ints = passing.findAll('td', {'data-stat': 'pass_int'})
            pass_ints = pass_ints[-1].get_text()
            
        except AttributeError:
            pass_ints = 0  
            
        # passer rating
        try:
            passer_rating = passing.findAll('td', {'data-stat': 'pass_rating'})
            passer_rating = passer_rating[-1].get_text()
            
        except AttributeError:
            passer_rating = 0    

        # Rushing_Receiving games
        try:
            rush_rec_games = rushing.findAll('td', {'data-stat': 'g'})
            rush_rec_games = rush_rec_games[-1].get_text()
            
        except AttributeError:
            rush_rec_games = 0
            
        # rushing attempts
        try:
            rushing_att = rushing.findAll('td', {'data-stat': 'rush_att'})
            rushing_att = rushing_att[-1].get_text()
            
        except AttributeError:
            rushing_att = 0
            
        # rushing yds
        try:
            rushing_yds = rushing.findAll('td', {'data-stat': 'rush_yds'})
            rushing_yds = rushing_yds[-1].get_text()
            
        except AttributeError:
            rushing_yds = 0
        
        # rushing yds per attempt
        try:
            rushing_yds_per_att = rushing.findAll('td', {'data-stat': 'rush_yds_per_att'})
            rushing_yds_per_att = rushing_yds_per_att[-1].get_text()
            
        except AttributeError:
            rushing_yds_per_att = 0
        
        # rushing tds
        try:
            rushing_tds = rushing.findAll('td', {'data-stat': 'rush_td'})
            rushing_tds = rushing_tds[-1].get_text()
            
        except AttributeError:
            rushing_tds = 0
        
        # receptions
        try:
            reception = rushing.findAll('td', {'data-stat': 'rec'})
            reception = reception[-1].get_text()
            
        except AttributeError:
            reception = 0
        
        # reception yds
        try:
            rec_yards = rushing.findAll('td', {'data-stat': 'rec_yds'})
            rec_yards = rec_yards[-1].get_text()
            
        except AttributeError:
            rec_yards = 0
        
        # yds per reception
        try:
            yds_per_reception = rushing.findAll('td', {'data-stat': 'rec_yds_per_rec'})
            yds_per_reception = yds_per_reception[-1].get_text()
            
        except AttributeError:
            yds_per_reception = 0
        
        # td receptions
        try:
            td_receptions = rushing.findAll('td', {'data-stat': 'rec_td'})
            td_receptions = td_receptions[-1].get_text()
            
        except AttributeError:
            td_receptions = 0
        
        # scrimmages
        try:
            scrimmages = rushing.findAll('td', {'data-stat': 'scrim_att'})
            scrimmages = scrimmages[-1].get_text()
            
        except AttributeError:
            scrimmages = 0
        
        # scrimmage yds
        try:
            scrimmage_yds = rushing.findAll('td', {'data-stat': 'scrim_yds'})
            scrimmage_yds = scrimmage_yds[-1].get_text()
            
        except AttributeError:
            scrimmage_yds = 0
        
        # scrimmage yds per attempt
        try:
            scrimmage_yds_per_att = rushing.findAll('td', {'data-stat': 'scrim_yds_per_att'})
            scrimmage_yds_per_att = scrimmage_yds_per_att[-1].get_text()
            
        except AttributeError:
            scrimmage_yds_per_att = 0
        
        # scrimmage tds
        try:
            scrimmage_tds = rushing.findAll('td', {'data-stat': 'scrim_td'})
            scrimmage_tds = scrimmage_tds[-1].get_text()
            
        except AttributeError:
            scrimmage_tds = 0
                
        # kicking games 
        try:
            kick_games = kicking.findAll('td', {'data-stat': 'g'})
            kick_games = kick_games[-1].get_text()
            
        except AttributeError:
            kick_games = 0
            
        # extra points 
        try:
            xp_made = kicking.findAll('td', {'data-stat': 'xpm'})
            xp_made = xp_made[-1].get_text()
            
        except AttributeError:
            xp_made = 0    
            
        # extra point attempts
        try:
            xp_attempt = kicking.findAll('td', {'data-stat': 'xpa'})
            xp_attempt = xp_attempt[-1].get_text()
            
        except AttributeError:
            xp_attempt = 0     
            
        # extra point percentage
        try:
            xp_perc = kicking.findAll('td', {'data-stat': 'xp_pct'})
            xp_perc = xp_perc[-1].get_text()
            
        except AttributeError:
            xp_perc = 0    
        
        # field goals
        try:
            field_goals = kicking.findAll('td', {'data-stat': 'fgm'})
            field_goals = field_goals[-1].get_text()
            
        except AttributeError:
            field_goals = 0 
        
        # field goal attempts
        try:
            field_goals_att = kicking.findAll('td', {'data-stat': 'fga'})
            field_goals_att = field_goals_att[-1].get_text()
            
        except AttributeError:
            field_goals_att = 0 
        
        # field goal percent
        try:
            fg_percent = kicking.findAll('td', {'data-stat': 'fg_pct'})
            fg_percent = fg_percent[-1].get_text()
            
        except AttributeError:
            fg_percent = 0
        
        # ttl kicking points
        try:
            ttl_kick_pts = kicking.findAll('td', {'data-stat': 'kick_points'})
            ttl_kick_pts = ttl_kick_pts[-1].get_text()
            
        except AttributeError:
            ttl_kick_pts = 0
        
        # num punts
        try:
            punts = kicking.findAll('td', {'data-stat': 'punt'})
            punts = punts[-1].get_text()
            
        except AttributeError:
            punts = 0
        
        # punt yards
        try:
            punt_yds = kicking.findAll('td', {'data-stat': 'punt_yds'})
            punt_yds = punt_yds[-1].get_text()
            
        except AttributeError:
            punt_yds = 0
        
        # yards per punt
        try:
            yds_per_punt = kicking.findAll('td', {'data-stat': 'punt_yds_per_punt'})
            yds_per_punt = yds_per_punt[-1].get_text()
            
        except AttributeError:
            yds_per_punt = 0
        
        # punt return games 
        try:
            puntret_games = punt_ret.findAll('td', {'data-stat': 'g'})
            puntret_games = puntret_games[-1].get_text()
        
        except AttributeError:
            puntret_games = 0
        
       # punts returned 
        try:
            ret_punts = punt_ret.findAll('td', {'data-stat': 'punt_ret'})
            ret_punts = ret_punts[-1].get_text()
        
        except AttributeError:
            ret_punts = 0 
        
        # returned yds
        try:
            ret_punts_yds = punt_ret.findAll('td', {'data-stat': 'punt_ret_yds'})
            ret_punts_yds = ret_punts_yds[-1].get_text()
        
        except AttributeError:
            ret_punts_yds = 0
        
        # yds per punt return
        try:
            yds_per_punt_return = punt_ret.findAll('td', {'data-stat': 'punt_ret_yds_per_ret'})
            yds_per_punt_return = yds_per_punt_return[-1].get_text()
        
        except AttributeError:
            yds_per_punt_return = 0
        
        # punt returned for a TD
        try:
            punt_ret_TD = punt_ret.findAll('td', {'data-stat': 'punt_ret_td'})
            punt_ret_TD = punt_ret_TD[-1].get_text()
        
        except AttributeError:
            punt_ret_TD = 0
        
        # kickoff returns
        try:
            kick_returns = punt_ret.findAll('td', {'data-stat': 'kick_ret'})
            kick_returns = kick_returns[-1].get_text()
        
        except AttributeError:
            kick_returns = 0
        
        # kickoff return yds
        try:
            kick_return_yds = punt_ret.findAll('td', {'data-stat': 'kick_ret_yds'})
            kick_return_yds = kick_return_yds[-1].get_text()
        
        except AttributeError:
            kick_return_yds = 0
        
        # yds per kickoff return
        try:
            yds_per_KO_ret = punt_ret.findAll('td', {'data-stat': 'kick_ret_yds_per_ret'})
            yds_per_KO_ret = yds_per_KO_ret[-1].get_text()
        
        except AttributeError:
            yds_per_KO_ret = 0
        
        # kickoff returned for a TD
        try:
            KO_ret_TD = punt_ret.findAll('td', {'data-stat': 'kick_ret_td'})
            KO_ret_TD = KO_ret_TD[-1].get_text()
        
        except AttributeError:
            KO_ret_TD = 0
        
        # scoring games 
        try:
            scoring_games = scoring.findAll('td', {'data-stat': 'g'})
            scoring_games = scoring_games[-1].get_text()
        
        except AttributeError:
            scoring_games = 0
        
        # other TDs scored 
        try:
            other_tds = scoring.findAll('td', {'data-stat': 'td_other'})
            other_tds = other_tds[-1].get_text()
        
        except AttributeError:
            other_tds = 0
        
        # other TDs scored 
        try:
            other_tds = scoring.findAll('td', {'data-stat': 'td_other'})
            other_tds = other_tds[-1].get_text()
        
        except AttributeError:
            other_tds = 0
        
        # ttl TDs scored 
        try:
            total_tds = scoring.findAll('td', {'data-stat': 'td_total'})
            total_tds = total_tds[-1].get_text()
        
        except AttributeError:
            total_tds = 0
       
        # 2 pt conversions 
        try:
            two_pt_conv = scoring.findAll('td', {'data-stat': 'two_pt_md'})
            two_pt_conv = two_pt_conv[-1].get_text()
        
        except AttributeError:
            two_pt_conv = 0
        
        # safety
        try:
            safety = scoring.findAll('td', {'data-stat': 'safety_md'})
            safety = safety[-1].get_text()
        
        except AttributeError:
            safety = 0
        
        # total points scored
        try:
            ttl_pts = scoring.findAll('td', {'data-stat': 'points'})
            ttl_pts = ttl_pts[-1].get_text()
        
        except AttributeError:
            ttl_pts = 0
        
        time.sleep(.5)
        
        global college_stats_df
        college_stats_df = college_stats_df.append({'Link': str(url),
                                                   'Defense_Games': defense_games, 
                                                   'Solo_Tackles': solotackles, 
                                                   'Assisted_Tackles': assistedtackles, 
                                                   'Ttl_Tackles': ttl_tackles, 
                                                   'Loss': tackle_loss, 
                                                   'Sacks': sacks, 
                                                   'Defensive_Interceptions': defense_int, 
                                                   'Def_Int_Yds': int_yards, 
                                                   'Yds_per_Int': yards_per_int, 
                                                   'Pick_6': pick6, 
                                                   'Defended_Passes': defended_pass, 
                                                   'Recovered_Fumbles': recovered_fumbles,
                                                   'Rec_Fumbles_Yds': fumble_recover_yrds, 
                                                   'Fumbles_Returned_TD': recovered_fumbles_td, 
                                                   'Forced_Fumbles': forced_fumbles,
                                                   'Passing_Games': passing_games, 
                                                   'Completions': completions,
                                                   'Pass_Attempts': pass_attempts,
                                                   'Completion_Percent': complete_rate,
                                                   'Pass_Yards': passing_yards,
                                                   'Pass_Yds_per_Attempt': yds_per_attempt,
                                                   'Adj_Pass_Yds_per_Attempt': adj_yds_per_attempt,
                                                   'Pass_TDs': pass_tds,
                                                   'Pass_Interceptions': pass_ints,
                                                   'Passer_Rating': passer_rating,
                                                   'Rushing/Receiving_Games': rush_rec_games,
                                                   'Rush_Attempts': rushing_att,
                                                   'Rush_Yds': rushing_yds,
                                                   'Rush_Yds_per_Attempt': rushing_yds_per_att, 
                                                   'Rush_TDs': rushing_tds,
                                                   'Receptions': reception,
                                                   'Rec_Yds': rec_yards,
                                                   'Rec_Yds_per_Reception': yds_per_reception,
                                                   'Rec_TDs': td_receptions,
                                                   'Plays_from_Scrimmage': scrimmages,
                                                   'Scrimmage_Yds': scrimmage_yds,
                                                   'Scrimmage_Yds_per_Attempt': scrimmage_yds_per_att,
                                                   'Scrimmage_TDs': scrimmage_tds,
                                                   'Kicking_Games': kick_games,
                                                   'XP_Made': xp_made,
                                                   'XP_Attempts': xp_attempt,
                                                   'XP_Percent': xp_perc,
                                                   'FG_Made': field_goals,
                                                   'FG_Attempts': field_goals_att,
                                                   'FG_Percent': fg_percent,
                                                   'TTL_Kicking_Points': ttl_kick_pts,
                                                   '#Punts': punts,
                                                   'Punt_Yds': punt_yds,
                                                   'Yds_per_Punt': yds_per_punt,
                                                   'PuntRet_Games': puntret_games,
                                                   'Punt_Returns': ret_punts,
                                                   'Punt_Return_Yds': ret_punts_yds,
                                                   'Yrds_per_Return': yds_per_punt_return,
                                                   'Punt_Returned_for_TD': punt_ret_TD,
                                                   'Kickoff_Returns': kick_returns,
                                                   'KO_Return_Yds': kick_return_yds,
                                                   'Yds_per_KO_Return': yds_per_KO_ret,
                                                   'KO_Returned_for_TD': KO_ret_TD,
                                                   'Scoring_Games': scoring_games,
                                                   'TD_Other': other_tds,
                                                   'Ttl_TDs': total_tds,
                                                   '2PT_Converstions': two_pt_conv,
                                                   'Safety': safety,
                                                   'TTL_Points': ttl_pts  
                                                   },
                                                  ignore_index=True)
```
